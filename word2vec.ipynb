```python
import gensim
import numpy as np
import pandas as pd
```

```python
# Load Google's pre-trained Word2Vec model.
model = gensim.models.KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True) 
```

```python
df = pd.read_pickle('./croll_exp/all_articles_df.p')
```

```python
def remove_none_in_list(li):
    li = [x for x in li if (x != 'Null' and x != '')]
    return li
```

```python
df.keywords = df.keywords.apply(remove_none_in_list)
```

```python
df = df[df.keywords.str.len() != 0]
df = df[df['keywords'] != False]
```

```python
df.to_pickle('remove_null_in_keywords.p')
```

```python
df.index = pd.RangeIndex(len(df.index))
df
```

```python
drop_indexes = []
for i in range(len(df.index)):
    if(df.loc[i,'abstract'] is None):
        drop_indexes.append(i)
```

```python
drop_indexes[:10]
```

```python
df = df.drop(drop_indexes)
```

```python
df.index = pd.RangeIndex(len(df.index))
df.head()
```

```python
df.to_pickle('remove_none_in_abs.p')
```
